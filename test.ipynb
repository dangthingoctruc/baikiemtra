```
#câu 1: cho ListA = [1,10,20,5,60,30,10,9,10], viết hàm List từ ListA bỏ đi các giá trị 10
listA = [1,10,20,5,60,30,10,9,10]
listB = []
for x in listA:
  if x!=10:
   listB.append(x)
print(listB)
```

```
#câu 2:cho 2 list 
#list1 = ["A","B","C","D","E","F"]
#list2 = [1,2,3,4,5,6]
list1=["A","B","C","D","E","F"]
list2=[1,2,3,4,5,6]
dic={}
for x in range(len(list1)):
	 dic[list1[x]] = list2[x]
print(dic)
```
